### Getting the right data

We want to plot the temperature for March 30 and March 31 only. In order to do this, we need to do the following:
    1. Convert the "timestamp" series to a format called "datetime"
    2. Change the index to timestamp

```python
# Series from a data frame are referenced using brackets: [ ]

data['timestamp'] = pd.to_datetime(data['timestamp'])
```

```python
data.index = data['timestamp']
```

### Lets see how the dataframe has changed

```python
data.head()
```

### Delete that extra timestamp column.

```python
del data['timestamp']
```

### Filtering the data set.

There are 3 temperature sensors. We just want to look at one (TMP421). Make a new dataframe that only has the temperature data form the TMP421.

```python
TMP421 = data[data['sensor'] == "TMP421"].value
```

```python
TMP421
```

### Select the date range

```python
TMP421_tempDay = TMP421['2018-03-30':'2018-03-31']
```

## Make the Plot

```python
plt.plot(TMP421_tempDay)
```

```python
plt.title("My First Data Vis")
my_plot = plt.gcf()
my_plot.set_size_inches(14, 7)
plt.ylabel("Temperature, C")
plt.xlabel("Time")
plt.title("My First Data Vis", fontsize = 20)
```
